```python
loan1 = loan.dropna(thresh=len(loan)*0.3, axis=1)
```

```python
### Convert data types

def convert_dtype(df, col, dtype):
    '''The function takes in a dataframe: df, a list of column names, and the desired data type'''
    for cname in col:
        df[cname] = df[cname].astype(dtype)
    
convert_dtype(loan1, ['term','grade','sub_grade','home_ownership','verification_status','purpose','title','application_type'],'category')    

```

```python
### Strings/Text extraction

#loan1.head()
```

```python
loan1['term'] = loan1['term'].str.extract('([0-9]+)')
```

```python
loan1['emp_length'] = loan1['emp_length'].str.extract('([0-9]+)')
```

```python
### Date time manipulation


# convert columns to date time dtype
def convert_dtype_datetime(df, col):
    for cname in col:
        df[cname] = pd.to_datetime(df[cname])
        
convert_dtype_datetime(loan1, ['issue_d','earliest_cr_line','last_pymnt_d','next_pymnt_d','last_credit_pull_d'])    


#### Alternatively, one could specify "parse_dates" parameter in read_csv: 
#pd.read_csv(file, sep='\t', header=None, names=headers, dtype=dtypes, parse_dates=parse_dates)
```
