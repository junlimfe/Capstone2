```python
import pandas as pd
loan = pd.read_csv("EDA_output.csv")
```

```python
#loan.describe().T
#loan.info()
```

```python
loan = loan.select_dtypes(exclude=['object'])
```

```python
loan_dummies = pd.get_dummies(loan,drop_first=True)
```

```python
loan_dummies.describe().T
```

```python
loan_dummies.info()
```

```python
loan_dummies.isna().sum().sum()
```

```python
loan_dummies.dropna(inplace=True)
```

```python
#loan_dummies.shape()
```

```python
loan_dummies_sample = loan_dummies.loc[:226066,]
```

```python
#loan_dummies_reserve.info()
```

```python
#scale data
from sklearn import preprocessing
import numpy as np
X = loan_dummies_sample.drop(['int_rate','Unnamed: 0', 'Unnamed: 0.1'], axis=1)
scaler = preprocessing.StandardScaler().fit(X)
X_scaled=scaler.transform(X) 
```

```python
y = loan_dummies_sample['int_rate']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=.8, random_state=1)
```

```python
#X_train
```

```python
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import f1_score
from sklearn.metrics import auc
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,roc_curve,roc_auc_score
from sklearn.metrics import accuracy_score,log_loss
from matplotlib import pyplot
```

```python
import time
from sklearn.ensemble import RandomForestRegressor
clf = RandomForestRegressor(n_estimators=300, random_state = 1,n_jobs=-1)

lab_enc = preprocessing.LabelEncoder()
training_scores_encoded = lab_enc.fit_transform(y_train)


model_res = clf.fit(X_train, training_scores_encoded)

y_pred = model_res.predict(X_test)

test_scores_encoded = lab_enc.fit_transform(y_test)

```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {'n_estimators':np.arange(100,301,100)}
rfr = RandomForestRegressor()
rfr_cv= GridSearchCV(rfr,param_grid,cv=4)

lab_enc = preprocessing.LabelEncoder()
training_scores_encoded = lab_enc.fit_transform(y_train)

rfr_cv.fit(X_train, training_scores_encoded)

```

```python
print("Best Parameters: " + str(rfr_cv.best_params_))
```
